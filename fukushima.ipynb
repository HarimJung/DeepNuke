```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

# data 2018

```python
data_dict = {1:31, 2:28, 3:31, 4:30, 5:31, 6:30, 7:31, 8:31, 9:30, 10:31, 11:30, 12:31}
data_list = []
for i in range(1, 13):
    a = data_dict[i]
    for j in range(1, a+1):
        source = "https://www4.tepco.co.jp/en/nu/fukushima-np/f2-rt/csv-e/f2-mp-2018%02d%02d23-e.csv" %(i, j)
        data_list.append(pd.read_csv(source))
```

```python
data_2018=pd.concat(data_list,axis=1)
```
